### Data plots for finding relation between missing value and SalePrice / Target Feature


```python
data= dataset.copy()

for feature in feature_with_nan:
    #make 1 if data is null else 0
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    # calculate mean sales price where info i smissing
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.ylabel('SalePrice')
    plt.xlabel('0 = Not Missing & 1 = Missing')
    plt.show()
```

We are finding realtionship between missing value and SalaPrice( Independant variable)

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
numerical_features
```

```python
print(len(numerical_features))
dataset[numerical_features].head()
```

```python
year_fearture = [feature for feature in numerical_features if 'yr' in feature or 'Year' in feature  or 'Yr' in feature]
```

```python
for feature in year_fearture:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.title(feature)
plt.ylabel('MedianPrice')
plt.xlabel('YearSold')
plt.show()
```

```python
year_fearture
```

```python
# Calculating Age Of House by YearSold
data = dataset.copy()
for feature in year_fearture:
    if feature != 'YrSold':
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel('Age = ' + feature + ' - YearSold')
        plt.ylabel('SalePrice')
        plt.show()
```

# Observation

```python
## we will be using logrithmic tranformation 
data = dataset.copy()
for feature in continues_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
  
        data[feature].hist()
        plt.title(feature)
        plt.ylabel('Count')
        plt.xlabel(feature)
        plt.show()
```

```python
## we will be using logrithmic tranformation 
data = dataset.copy()
for feature in continues_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        plt.scatter(data[feature], data['SalePrice'])
        plt.ylabel('SalePrice')
        plt.xlabel(feature)
        plt.show()
```

## Outliers Detection 


```python
for feature in continues_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]= np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
